#**Functions in Python**

A function in Python is a reusable block of code that performs a specific task. It helps in organizing code, improving readability, and reducing redundancy. Functions allow for code reuse, making programs more modular and maintainable. They are defined using the def keyword, followed by a function name, parentheses, and a colon. The code block within the function is indented.



**Use Cases of Functions**
- **Code Reusability:** Functions allow you to reuse code without rewriting it multiple times.
- **Modularity:** Functions enable breaking down complex problems into smaller, manageable pieces.
- **Maintainability:** Changes can be made to a function's code without affecting the rest of the program.
- **Abstraction:** Functions can hide the implementation details, allowing users to use them without understanding the underlying code.



**Advantages of Functions**

- **Increases Modularity:** Python functions help divide the code into smaller problems and solve them individually, thus making it easier to code.

- **Minimizes Redundancy:** Python functions help you save the effort of rewriting the whole code. All you got to do is call the Function once it is defined.

- **Maximizes Reusability:** Once defined, the python functions can be called as many times as needed, thus enhancing code reusability.

- **Helps in Debugging:** As the large program is divided into sections with the help of functions, it helps in increasing the readability of code, at the same time ensuring easy debugging.

#**Syntax of Functions**

###**1. Defining a Function:**

- To create a function, you start with the def keyword followed by the function name and parentheses ( ). This is where you can specify if the function needs any inputs.

```
def my_function(input1, input2):
```



**The syntax for defining a function is as follows:**

````python
def function_name(parameters):
    """Docstring explaining the function."""
    # Function body
    return value
````

- **def:** A keyword to start the function definition.
- **function_name:** A unique name for the function.
- **parameters:** Optional input values that the function can accept.
- **return:** A statement to return a value from the function.

###**2. Function Body:**

- The function body consists of the code block that defines what the function does. It's indented and follows immediately after the function declaration.



```
def my_function(name):
  print(name)
```

- To call a function, use its name followed by parentheses.
- If the function requires arguments, pass them inside the parentheses.



**Example:**

````python
# Define a function named 'add' with two parameters
def add(a, b):
    return a + b

# Call the function with arguments
result = add(2, 3)

# Output the result
print(result)  # Output: 5
````

- The function add is defined with two parameters, a and b.
- The function is called with the arguments 2 and 3.
- The result of the function call is stored in the variable result.
- The result is then printed, which outputs 5

###**3. Return Statement:**

- Not all functions need to return a value, but if they do, it's done using the return keyword. This statement ends the function’s execution and passes back a result to the caller.

```
def my_function(input1, input2):
    result = input1 + input2
    return result
```

**Example:**

````python
# Define a function named 'add_numbers' that takes two parameters
def add_numbers(a, b):
    # Calculate the sum of a and b
    sum = a + b

    # Return the calculated sum
    return sum

# Call the function with arguments 7 and 3, and store the result in 'result'
result = add_numbers(7, 3)

# Output the result
print(result)  # Output: 10
````

**Explanation:**

Function Definition (add_numbers):
- The function takes two parameters, a and b.
- Inside the function, the sum of a and b is calculated and stored in the variable sum.
- The return sum statement sends the value of sum back to the caller.

Function Call:
- The function add_numbers is called with the arguments 7 and 3.
- The function computes the sum as 7 + 3 = 10 and returns this value.

Output:
- The returned value 10 is stored in the variable result.
- The print(result) statement outputs the value of result, which is 10.

###**4. Function Call:**

- To execute a function, you simply write its name followed by parentheses ( ). If the function requires inputs, you pass them inside these parentheses.
```
result = my_function(5, 3)  # Calling the function with arguments 5 and 3
```

#**Variable Characterization Based on Functions**

image.png

###**1. Local Variables:**
- A variable declared inside the Function’s body is called a local variable. Such a variable is said to have a local scope, i.e., it is only accessible within the Function and not outside.

````python
def example():
    local_var = 10
    print(local_var)  # Output: 10
````

###**2. Global Variables:**
- Any variable declared outside the Function is called a global variable. Such a variable has global scope, i.e., it is accessible everywhere, whether inside or outside of the Function.

````python
global_var = 20

def example():
    print(global_var)  # Output: 20
````

###**3. Nonlocal Variables:**
- Variables defined in the nearest enclosing scope, not the global scope.

````python
def outer():
    x = 5

    def inner():
        nonlocal x
        x = 10
    inner()
    print(x)  # Output: 10
````

#**Parameters and Arguments**



**Parameters:** Parameters are the variables listed inside the parentheses in the function definition. They act as placeholders for the data that will be passed into the function when it is called.

**Arguments:**  Arguments, on the other hand, are the actual values or variables that are passed to the function when it is called. They match the parameters defined in the function signature.

image.png

**Types of Parameters:**
###**1. Positional Parameters:**
- Parameters that need to be in the correct order.

````python
def add(a, b):
    return a + b

add(2, 3)  # Positional arguments: 2 and 3
````

###**2. Keyword Parameters:**
- Parameters passed by explicitly naming the parameter.

````python
def greet(name, message):
    print(f"Hello {name}, {message}")

greet(name="Alice", message="Good morning!")  # Output: Hello Alice, Good morning!
````

###**3. Default Parameters:**
- Parameters that have a default value if no argument is passed.

````python
def greet(name="Guest"):
    print(f"Hello {name}")

greet()  # Output: Hello Guest
````

###**4. Variable-length Parameters:**
- Parameters that accept an arbitrary number of arguments.

````python
def add(*numbers):
    return sum(numbers)

add(1, 2, 3, 4)  # Output: 10
````

###**3. Lambda Functions**

- Lambda functions are small, unnamed functions defined using the lambda keyword.
- They are typically used for simple operations and are often employed when a function is required for a short period and doesn't need a formal name. The general syntax of a lambda function is as follows:

```python
lambda arguments: expression
```

Lambda functions have the following characteristics:

- They can take any number of arguments but can contain only a single expression.
- They are concise and primarily used for simple operations.

image.png

**Syntax**

````python
lambda parameters: expression
````

**Example:**

````python
square = lambda x: x * x
print(square(5))  # Output: 25
````
