##**Introduction to Strings**

- A string in Python is a sequence of characters enclosed within single ('), double ("), or triple quotes (''' or """). Strings are one of the most commonly used data types in Python, and they are immutable, meaning once a string is created, it cannot be modified.

**Why Strings Are Important**
- Strings are fundamental in programming as they are used for storing and manipulating text. From simple text messages to complex data processing, strings are an essential part of any programming language, including Python.

##**Creating Strings in Python**

- Strings are sequences of characters used to store and manipulate text. In Python, creating strings is straightforward, and they are one of the most commonly used data types. Let's explore how to create strings and some of their basic features.

#### **Single Quotes**

- You can create a string using single quotes (').

```python
# Creating a string with single quotes
single_quote_string = 'Hello, World!'
print(single_quote_string)  # Output: Hello, World!
```

#### **Double Quotes**

- Strings can also be created using double quotes ("). This is useful if your string contains a single quote character.

```python
# Creating a string with double quotes
double_quote_string = "Python is fun!"
print(double_quote_string)  # Output: Python is fun!
```

- Triple quotes (''' or """) are used to create multi-line strings or strings that contain both single and double quotes.

```python
# Creating a multi-line string with triple quotes
multi_line_string = """This is a string
that spans multiple lines."""
print(multi_line_string)
# Output:
# This is a string
# that spans multiple lines.
```

#### **String Containing Quotes**

- If your string contains both single and double quotes, you can use triple quotes to avoid syntax errors.

##**Accessing Characters in a String**

- In Python, strings are sequences of characters, and each character in a string has a position, known as an index. You can access individual characters in a string using their index. Let’s explore how to access characters in a string.

**1. Indexing**
- Python uses zero-based indexing, meaning the first character in a string is at index 0, the second character is at index 1, and so on.

school.jpg

```python
# Example string
greeting = "SCHOOL"

# Accessing characters by index
print(greeting[0])  # Output: S (first character)
print(greeting[1])  # Output: C (second character)
print(greeting[4])  # Output: O (fifth character)
```

**2. Negative Indexing**
- Python also supports negative indexing, where the last character in a string has an index of -1, the second-to-last character is -2, and so on.

```python
# Accessing characters using negative indexing
print(greeting[-1])  # Output: L (last character)
print(greeting[-2])  # Output: O (second-to-last character)
```

**3. Slicing**
- You can access a range of characters (a substring) by using slicing. Slicing is done by specifying a start index and an end index, separated by a colon :. The slice includes the start index but excludes the end index.

```python
# Slicing a string
print(greeting[1:4])  # Output: CHO (characters from index 1 to 3)
```

- **Start Index:** The index of the first character in the slice.
- **End Index:** The index right after the last character in the slice.

**4. Slicing with Defaults**
- If you omit the start index, Python starts the slice from the beginning of the string. If you omit the end index, Python continues the slice to the end of the string.

```python
# Slicing from the beginning to a specific index
print(greeting[:3])  # Output: SCH (first three characters)

# Slicing from a specific index to the end
print(greeting[2:])  # Output: HOOL (characters from index 2 to the end)
```

**5. Slicing with Step**
- You can specify a step in the slicing process, which determines how many characters to skip between each character in the slice.

```python
# Slicing with a step
print(greeting[0:5:2])  # Output: SHO (characters at index 0, 2, and 4)
```

- **Step:** The number of characters to skip between each character in the slice.

**6. String Immutability**
- It’s important to note that strings in Python are immutable, meaning once a string is created, it cannot be modified. You can’t change a character in a string directly by assigning a new value to it.

- If you need to change a string, you must create a new string.

```python
# Creating a new string with a modification
new_greeting = 'T' + greeting[1:]
print(new_greeting)  # Output: TCHOOL
```

##**String Operations in Python**
- Python provides a wide range of operations that you can perform on strings. These operations allow you to manipulate, combine, and analyze text data easily. Here are some of the most common string operations in Python:

Screenshot 2024-08-22 at 10.59.08 AM.png

**1. Concatenation**
- Concatenation is the process of joining two or more strings together using the + operator.

```python
# Concatenating strings
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)  # Output: John Doe
```

**2. Repetition**
- You can repeat a string multiple times using the * operator.

```python
# Repeating a string
laugh = "ha"
laugh_repeated = laugh * 3
print(laugh_repeated)  # Output: hahaha
```

**3. String Length**
- The len() function returns the length of a string, which is the number of characters it contains.

```python
# Finding the length of a string
message = "Hello, World!"
length = len(message)
print(length)  # Output: 13
```

**4. String Containment**
- The in and not in operators are used to check if a substring is present in a string.

```python
# Checking if a substring is in a string
message = "Welcome to Python"
print("Python" in message)  # Output: True
print("Java" not in message)  # Output: True
```

**5. Strip Characters**
- You can remove leading and trailing whitespace (or other specified characters) from a string using the strip(), lstrip(), and rstrip() methods.

```python
# Stripping whitespace
text = "   Hello, World!   "
print(text.strip())  # Output: Hello, World!
print(text.lstrip()) # Output: Hello, World!
print(text.rstrip()) # Output:    Hello, World!
```

**6. Find and Replace Substrings**
- **find(substring):** Returns the index of the first occurrence of the substring, or -1 if not found.
- **replace(old, new):** Replaces all occurrences of the old substring with the new substring.

```python
# Finding and replacing substrings
message = "Hello, World!"
print(message.find("World"))    # Output: 7
print(message.replace("World", "Python"))  # Output: Hello, Python!
```

**7. Splitting and Joining Strings**
- **split(separator):** Splits the string into a list of - - substrings based on the specified separator.
- **join(iterable):** Joins the elements of an iterable (e.g., list) into a single string, with each element separated by the specified separator.

```python
# Splitting and joining strings
sentence = "This is a sample sentence."
words = sentence.split()  # Default separator is whitespace
print(words)  # Output: ['This', 'is', 'a', 'sample', 'sentence.']

# Joining list elements into a string
joined_sentence = " ".join(words)
print(joined_sentence)  # Output: This is a sample sentence.
```

##**String Formatting in Python**
- String formatting in Python allows you to create formatted strings by embedding variables and expressions into strings. This is useful for creating dynamic output and for making your code more readable and maintainable. Python provides several ways to format strings:

How-To-Format-A-String-in-python.jpg

**1. Old-Style String Formatting (% Operator)**
- This method uses the % operator to embed variables into a string. It is an older method but still commonly used.

```python
name = "Alice"
age = 30
formatted_string = "My name is %s and I am %d years old." % (name, age)
print(formatted_string)  # Output: My name is Alice and I am 30 years old.
```

- %s is used for strings.
- %d is used for integers.
- %f is used for floating-point numbers (you can specify the number of decimal places with %.2f, for example).

**2. str.format() Method**
- The str.format() method is more powerful and flexible than the % operator. It allows you to use placeholders (curly braces {}) in the string, which are replaced by the values passed to format().

```python
name = "Alice"
age = 30
formatted_string = "My name is {} and I am {} years old.".format(name, age)
print(formatted_string)  # Output: My name is Alice and I am 30 years old.
```

**3. F-Strings (Formatted String Literals)**
- Introduced in Python 3.6, f-strings provide a concise and readable way to embed expressions inside string literals, using curly braces {} and prefixed with the letter f.

```python
name = "Alice"
age = 30
formatted_string = f"My name is {name} and I am {age} years old."
print(formatted_string)  # Output: My name is Alice and I am 30 years old.
```

- F-strings also allow you to embed expressions and perform calculations directly within the string:

```python
formatted_string = f"Next year, I will be {age + 1} years old."
print(formatted_string)  # Output: Next year, I will be 31 years old.
```

**4. String Interpolation with Template Class**
- The Template class from the string module allows for simple string substitutions, especially useful when the format string might be provided by the user.

```python
from string import Template

template = Template("My name is $name and I am $age years old.")
formatted_string = template.substitute(name="Alice", age=30)
print(formatted_string)  # Output: My name is Alice and I am 30 years old.
```

##**Immutability of Strings in Python**
- In Python, strings are immutable. This means that once a string is created, it cannot be changed. Any operation that appears to modify a string will actually create a new string rather than altering the original one. Let's explore what this means and why it's important.

**Understanding Immutability**
- When you create a string in Python, it's stored in memory, and its value cannot be changed. If you try to modify the string, Python will create a new string with the modified content, leaving the original string unchanged.



**For example:**

```python
# Original string
text = "Hello"

# Create a new string with the first character changed to 'h'
text = "h" + text[1:]

# Output the modified string
print(text)  # Output: "hello"
```

When the error occurs because strings are immutable, so you can't change them directly.

- What Happens When You "Modify" a String?
- When you perform an operation that seems to change a string, a new string is created and the original string remains unchanged.

**For example:**

```python
text = "Hello"
new_text = text.lower()  # This creates a new string "hello" and assigns it to new_text
```

Here, new_text will contain the string "hello", but text will still be "Hello". The lower() method doesn't change the original string; it returns a new one.



**Why Are Strings Immutable?**
- **Efficiency:** Immutability allows Python to optimize memory usage by reusing string objects. If two variables contain the same string, Python can point them to the same memory location rather than creating two separate objects.

- **Security:** Since strings are commonly used as keys in dictionaries and as elements in sets, immutability ensures that they cannot be altered after being added to these collections, which could lead to inconsistencies or errors.

- **Simplicity:** Immutable objects are inherently thread-safe, as they cannot be modified by different parts of a program simultaneously. This reduces the complexity of managing shared data in concurrent programming.


**Example 1: Assigning a New Value**

```python
text = "Hello"
text = "World"  # A new string "World" is created and assigned to text
```

- Here, the original string "Hello" is not modified. Instead, text is now pointing to a completely new string "World".

**Example 2: Concatenation**

```python
text = "Hello"
text += " World"  # Creates a new string "Hello World"
print(text)  # Output: Hello World
```

- Even though it looks like we're adding to text, a new string is created and assigned to text, replacing the reference to the original string.

##**Escape Characters in Python**


maxresdefault.jpg

- Escape characters in Python are special characters used in string literals that allow you to include characters that are otherwise difficult or impossible to include directly in a string. They are introduced with a backslash (\), followed by the character you want to escape or represent.

**Common Escape Characters**
- Here are some of the most commonly used escape characters in Python:

**1. Newline (\n)**

- Inserts a new line in the string.


```python
text = "Hello\nWorld"
print(text)

```

**2. Tab (\t)**

- Inserts a horizontal tab (indentation) in the string.

```python
text = "Hello\tWorld"
print(text)
```

**3. Backslash (\\)**

- Inserts a literal backslash in the string.

```python
text = "This is a backslash: \\"
print(text)
```

**4. Single Quote (\')**

- Allows you to include a single quote in a string that is enclosed in single quotes.

```python
text = 'It\'s a sunny day'
print(text)
```

**5. Double Quote (\")**

Allows you to include a double quote in a string that is enclosed in double quotes.

```python
text = "He said, \"Hello!\""
print(text)
```

**6. Carriage Return (\r)**

- Moves the cursor to the beginning of the line without advancing to the next line.

```python
text = "Hello\rWorld"
print(text)
```

**7. Backspace (\b)**

- Moves the cursor one position back, effectively removing the character before it.

```python
text = "Hello\b World"
print(text)
```

**8. Form Feed (\f)**

- Moves the cursor to the next line and inserts a form feed character.

```python
text = "Hello\fWorld"
print(text)
```

##**Raw Strings**
- Raw strings in Python are a way to define string literals where backslashes (\) are treated as literal characters and not as escape characters.
- This is particularly useful when dealing with strings that contain many backslashes, such as file paths, regular expressions, or any text where you don't want escape sequences to be processed.


**How to Create a Raw String**
- To create a raw string, you prefix the string literal with an r or R. This tells Python to treat backslashes in the string as literal backslashes, without any special meaning.

```python
# Regular string
path = "C:\\Users\\Name\\Folder"
print(path)  # Output: C:\Users\Name\Folder

# Raw string
raw_path = r"C:\Users\Name\Folder"
print(raw_path)  # Output: C:\Users\Name\Folder
```

- In the example above, both path and raw_path will output the same string, but raw_path was easier to write because you didn't need to double the backslashes.

**When to Use Raw Strings**
- **File Paths**
  - When dealing with Windows file paths, raw strings help avoid the need to escape backslashes.

```python
path = r"C:\new_folder\file.txt"
```

- **Regular Expressions**
  - Regular expressions often include backslashes, and using raw strings makes them easier to read and write.

```python
import re
pattern = r"\d+\.\d+"
result = re.findall(pattern, "The price is 12.99")
print(result)  # Output: ['12.99']
```

**Text with Many Backslashes**
- If you're working with text that contains a lot of backslashes, such as LaTeX commands or Markdown links, raw strings simplify the syntax.

```python
latex_command = r"\frac{a}{b}"
```

##**String Iteration in Python**
- String iteration refers to the process of looping through each character in a string, one by one. In Python, strings are sequences of characters, which means they can be iterated over using loops, typically for loops.

**Iterating Over a String**
- The most common way to iterate over a string is by using a for loop.

**Here's a basic example:**

```python
text = "Python"
for char in text:
    print(char)
```

- In this example, the loop goes through each character in the string text and prints it.

**Iterating with Indexes**
- Sometimes, you may need to access both the index and the character during iteration. You can achieve this using the enumerate() function:

```python
text = "Python"
for index, char in enumerate(text):
    print(f"Index {index}: {char}")
```

- Here, enumerate() provides a counter (starting from 0) along with the character in each iteration.

###**Iterating in Reverse**
- To iterate over a string in reverse order, you can use Python's slicing feature or the reversed() function.



**1. Using Slicing:**

```python
text = "Python"
for char in text[::-1]:
    print(char)
```

**2. Using reversed():**

```python
text = "Python"
for char in reversed(text):
    print(char)
```

###**Using a While Loop for Iteration**
- While less common, you can also iterate over a string using a while loop, particularly if you need more control over the loop's behavior.

```python
text = "Python"
index = 0
while index < len(text):
    print(text[index])
    index += 1
```

###**String Iteration with Conditions**
- You can combine iteration with conditional statements to perform operations only on certain characters.

```python
text = "Python Programming"
for char in text:
    if char in "aeiou":
        print(f"Vowel found: {char}")
```

###**Iterating Over Substrings**
- You might want to iterate over the string in chunks (substrings) rather than one character at a time. You can achieve this by adjusting the range or using slicing in your loop.

```python
text = "PythonProgramming"
chunk_size = 3
for i in range(0, len(text), chunk_size):
    print(text[i:i+chunk_size])
```

##**String Comparison in Python**
- String comparison in Python involves checking whether two strings are equal, not equal, or comparing their lexicographical order. Python provides several built-in operators and methods for comparing strings.

Screenshot 2024-08-22 at 12.38.39 PM.png

**1. Equality and Inequality**
- Equal (==): Checks if two strings have the same value.
Not Equal (!=): Checks if two strings have different values.

**Examples:**

```python
str1 = "Hello"
str2 = "Hello"
str3 = "World"

print(str1 == str2)  # Output: True
print(str1 != str3)  # Output: True
```

**2. Lexicographical Comparison**
- Strings are compared lexicographically based on their Unicode values. This means the comparison is based on the alphabetical order of characters, similar to how words are ordered in a dictionary.

- **Less Than (<):** Checks if the first string comes before the second string.
- **Greater Than (>):** Checks if the first string comes after the second string.
- **Less Than or Equal To (<=):** Checks if the first string comes before or is equal to the second string.
- **Greater Than or Equal To (>=):** Checks if the first string comes after or is equal to the second string.


**Examples:**

```python
str1 = "Apple"
str2 = "Banana"
str3 = "Apple"

print(str1 < str2)   # Output: True (because 'Apple' comes before 'Banana')
print(str1 > str2)   # Output: False
print(str1 <= str3)  # Output: True (because 'Apple' is equal to 'Apple')
print(str1 >= str3)  # Output: True
```

**3. Case Sensitivity**
- String comparisons are case-sensitive by default. This means that uppercase and lowercase characters are considered different.

**Examples:**

```python
str1 = "hello"
str2 = "Hello"

print(str1 == str2)  # Output: False
print(str1 > str2)   # Output: True (because lowercase 'h' has a higher Unicode value than uppercase 'H')
```

**4. Ignoring Case Sensitivity**
- To perform case-insensitive comparisons, you can convert both strings to the same case using .lower() or .upper() methods before comparing.


**Examples:**

```python
str1 = "Hello"
str2 = "hello"

print(str1.lower() == str2.lower())  # Output: True
print(str1.upper() == str2.upper())  # Output: True
```
