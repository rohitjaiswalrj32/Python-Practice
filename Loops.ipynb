```python
#Loops: Loops in Python are used to execute a block of code repeatedly as long as a condition is met.

        #There are two main types of loops in Python:
#       1. For Loop: Used to iterate over a sequence such as a list, string, tuple, or a range of numbers.
#       2. While Loop: Repeats a block of code while a condition is true.
```

```python
#Practice Questions - For Loop

#Q1. Print numbers from 10 to 1 (reverse order).

for i in range(10, 0, -1):
  print(i)

#Q2. Print all even numbers from 1 to 20.

for i in range(2, 20, 2):
  print(i)

#Q3. Print characters of your name one by one.

name = "RohitJaiswal"
for char in name:
  print(char)

#Q4. Print the table of any number (e.g., table of 5).

for i in range(5, 51, 5):
  print(i)

#Q5. Calculate the sum of numbers from 1 to N.

N = int(input("Enter a number: "))
sum = 0
for i in range(1, N + 1):
  sum += i
print("Sum of numbers from 1 to", N, "is:", sum)



```

```python
#Practice Questions - While Loop


#Q1. Print numbers from 1 to 100.

i = 1
while i <= 100:
  print(i)
  i += 1


#Q2. Print all odd numbers between 50 and 70.

i = 50
while i >= 50 and i <= 70:
  if i % 2 != 0:
    print(i)
  i += 1


#Q3. Calculate the factorial of a number.

factorial = 1
num = int(input("Enter a number: "))
i = 1
while i <= num:
  factorial *= i
  i += 1
  print("Factorial of", num, "is:", factorial)


#Q4. Reverse a number (e.g., input: 123 → output: 321).

num = int(input("Enter a number: "))
reversed_num = 0
while num > 0:
  digit = num % 10
  reversed_num = reversed_num * 10 + digit
  num = num // 10
print("Reversed number is:", reversed_num)

#Q5. Keep asking the user for input until they type "exit".

while True:
  user_input = input("Enter something (type 'exit' to quit): ")
  if user_input == "exit":
    break
```

```python
#Practice Questions – break & continue


# Q1. Print numbers 1 to 10, but stop when the number is 7.
for i in range(1, 11):
    if i == 7:
        break
    print(i)


# Q2. Print numbers 1 to 10, but skip 4 and 6.
for i in range(1, 11):
    if i == 4 or i == 6:
        continue
    print(i)


# Q3. Keep asking for numbers, break if the number is negative.
while True:
    number = int(input("Enter a number: "))
    if number < 0:
        print("You entered a negative number")
        break
    else:
        print("You entered:", number)
```
